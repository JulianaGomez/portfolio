```python
from transformers import AutoTokenizer # BPE - RoBERTa
from transformers import BertTokenizer # WordPiece - BERT
from mosestokenizer import * # Moses tokenizer - eLMo
```

## Define the three tokenizers

- Byte-pair encoding (BPE) is used in RoBERTa
- WordPiece is used in BERT
- Moses Tokenizer is used in eLMo

```python
checkpoint = 'bert-base-cased'
bert_tokenizer = BertTokenizer.from_pretrained(checkpoint)
```

```python
bpe_tokenizer = AutoTokenizer.from_pretrained("gpt2")
```

```python
moses_tokenizer = MosesTokenizer('en')
```

## Tokenize!

```python
bert_tokenizer.tokenize('This tree is 1253 years old.')
```

```python
bpe_tokenizer.tokenize('This tree is 1253 years old.')
```

```python
moses_tokenizer('This tree is 1253 years old.')
#tokenize.close()
```
